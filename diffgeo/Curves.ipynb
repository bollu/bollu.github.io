## Parametrized differentiable curve
Let $\alpha: (a, b) \to \mathbb R$. Then $\alpha$ is a **parametrized differentiable curve** iff $\alpha$ is differentiable in the domain.

## Tangent line
- Let $\alpha: (a, b) \to \mathbb R$.
- For each $t \in I$ such that $\alpha'(t) \neq 0$, there is a unique straight line which contains the point $\alpha(t)$ and the vector $\alpha'(t)$. This line is the tangent line to $\alpha$ at $t$.

## Regular curve
- A parametrized differentiable curve $\alpha: I \to \mathbb R$ is said to be **regular** iff $\alpha'(t) \neq 0$ for all $t \in I$.

## Arc length
- Given $t \in I$, the arc length from the point $t_0$ is defined to be:

$$
s(t) \equiv \int_{t_0}^t |\alpha'(t)|dt
$$

- where $|\alpha'(t)| = \sqrt{x'(t)^2 + y'(t)^2 + z'(t)^2}$. We have $ds/dt = |\alpha'(t)|$ (by fundamental theorem of integration: $F(t) = \int_{t_0}^t f(x) dx$ implies $F'(t) = f(t) - f(t_0)$].
- If the parameter $t$ is already the arclength measured from some basepoint $t_0$, then $ds/dt = ds/ds = 1 = |\alpha'(t)|$. So the velocity vector has length 1.
- On the other hand, if $|\alpha'(t)| = 1$, then $s(t) = \int_{t_0}^t |\alpha'(t)| dt = \int_{t_0}^t 1 dt = t - t_0$. That is, $t$ is the arc-length measured from some point.
- This means that we have an equivalence: $|\alpha'(t)| = 1$ iff $\alpha$ is arc-length parametrized.
- We shall restrict ourselves to arclength-parametrized curves, since we can alwas reparametrize curves to be arclength parametrized (TODO: why? Section 1.5)

## Change of orientation
- Define a curve by $\beta(s) \equiv \alpha(-s)$. This travels in the opposite direction --- this is a *change of orientation*.

## Orientation for vector space basis $V$
- Define two bases $\{ e_i \}$ and $\{ f_i \}$ to have the same orientation iff the change of basis matrix has positive determinant.
- This is an equivalence relation on the set of ordered bases of $V$
- If we choose some "basepoint" orientation $\{ b_i \}$, we get two equivalence classes, one with the same orientation as $b_i$, and one with the opposite one. 
- There is a natural orientation given by setting $\{ (1, 0, 0), (0, 1, 0), (0, 0, 1) \}$ to have orientation $+1$.

## Vector product (wedge product) / (cross product)
- Let $u, v \in \mathbb R^3$. The vector product/wedge product of $u, v$ (in that order) is defined by the relation $(u \wedge v) \cdot w) = det(u, v, w)$ for all $w \in \mathbb R^3$.
- Theorem: Vector product produces positive basis

# Local theory of arclength parametrized curves
- Let $\alpha$ be arclength parametrized. Then $\kappa_\alpha(s) \equiv |\alpha''(s)|$ is the curvature of $\alpha$.
- Easy to see that $\kappa_\alpha = 0$ iff $\alpha$ is a straight line.

## Change of orientation preserves curvature
- If $\beta(s) = \alpha(-s)$, then:

$$
\begin{align*}
&\beta'(s) \\
&= \frac{d\alpha(-s}{ds} \\
&= \frac{d\alpha(-s}{d(-s)} \frac{d(-s)}{ds} \\
&= \alpha'(-s) (-1)
&= - \alpha'(-s)
\end{align*}
$$

- This implies that the second derivative/curvature is $\beta''(s) = \alpha''(-s)$ due to another $-1$ sign when differentiating $\d\alpha'(-s)/ds$

## Normal vector to curve
- At locations where $\alpha''(s) \neq 0$, then we define the unit normal as $\hat n(s) \equiv \alpha''(s)/|\alpha''(s)|$ which is $\hat n(s) \equiv \alpha''(s)/\kappa(s)$.
- This implies that we can write $\alpha''(s) = \kappa(s) \hat n(s)$.

## Normal vector is normal to tangent vector
- We will prove that $\alpha''(s) \cdot \alpha(s) = 0$ when we have an arclengt parametrized curve $\alpha(s)$.
- See that $\alpha'(s) \cdot \alpha'(s) = 1$ since $\alpha'(s)$ is a unit vector.
- Differentiating, we get:

\begin{align*}
&\frac{d \alpha'(s) \cdot \alpha'(s)}{ds}  \\
&= \alpha''(s) \cdot \alpha'(s) + \alpha'(s) \cdot \alpha''(s) = 0 \\
&= 2 \alpha''(s) \cdot \alpha'(s) = 0 \\
\alpha''(s) \cdot \alpha'(s) = 0
\end{align*}

- Thus the normal vector is always normal/perpenducilar to the tangent vector.

## Osculating plane
- The plane containing $\alpha'(s)$ and $\alpha''(s)$ is the osculating plane at $s$.
- At points where $\kappa(s) = 0$, we don't have an osculating plane since $\hat n = 0$. 

## Singular point of order 1 / 0
- We need the osculating plane to do basically anything.
- So, we define a point with $\alpha''(s) = 0$ as a **singular point of order $1$**.
- Points where $\alpha'(s) = 0$ are singular points of order $0$.

## $t(s) \equiv \alpha'(s)$
- Define $t(s)$ as the tangent to the curve at $s$.
- Then we have $t'(s) = \alpha''(s) = \kappa(s) \hat n(s)$.

## Binormal vector
- Define $b(s) \equiv t(s) \wedge \hat n(s)$. This is a **unit vector** because $t(s)$ is a unit vector since we have an arclength parametrization, and $\hat n(s)$ is normal to $t(s)$ as we have proven.

## Derivative of Binormal vector $b'(s)$

- Consider $b'(s)$. This tells us how the vector $b(s)$ changes.
- Since $|b(s)| = 1$, $b'(s)$ does not record anything about change in length; rather, it records change in direction of $b(s)$.

#### Computing $b'(s)$

\begin{align*}
&b(s) = t(s) \wedge \hat n(s) \\
&b(s) \cdot t(s) = (t(s) \wedge \hat n(s)) \cdot t(s) = 0 \\
&(b(s) \cdot t(s))' = = 0 \\
&b'(s) \cdot t(s) + b(s) \wedge t'(s) = 0 \\
&b'(s) \cdot t(s) + b(s) \wedge \hat n(s) = 0 \\
&b'(s) \cdot t(s) + 0 = 0 \\
\end{align*}

-Thus $b'(s)$ is normal to $t(s)$.
- $b'(s)$ is normal to $b(s)$ since $b(s)$ is a unit vector.
- Thus $b'(s)$ is along $n(s)$ since $\{t, n, b \}$ form an orthonormal frame.

## Torsion of a curve $\tau(s)$
- Define $tau(s)$ by the relation $b'(s) = \tau(s) \hat n(s)$.
- Said differently, $\tau(s) = |b'(s)|$.
